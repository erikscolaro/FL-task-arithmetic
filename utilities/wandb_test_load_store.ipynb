```python
import torch
import wandb
from fl_task_arithmetic.task import Net
from utilities.wandb_utils import load_model_from_wandb, save_model_to_wandb

entity = "aml-fl-project"
project = "fl-task-arithmetic"
group = "experiment-0"
```

```python
# ------------------------------
# 1. Create the model
# ------------------------------
global_model = Net()

# Simulate that the model has non-zero weights
for p in global_model.parameters():
    p.data += 1.235

# ------------------------------
# 2. STORE: save the model as an artifact
# ------------------------------
run = wandb.init(
    entity=entity,
    project=project,
    group=group,
    name="server-store",
    id="server-store",
    resume="allow",
    mode="online",
)

save_model_to_wandb(
    run=run,
    model=global_model,
)

print("\n--- MODEL SAVED ON WANDB ---\n")
```

```python
# ------------------------------
# 3. LOAD: create an empty model
# ------------------------------
from pathlib import Path
import wandb
import torch
from torch.nn import Module
from pathlib import Path


loaded_model = Net()

# ------------------------------
# 4. LOAD: retrieve model from artifact
# ------------------------------

load_model_from_wandb(run=run, model=loaded_model)

print(loaded_model)

print("\n--- MODEL LOADED FROM WANDB ---\n")
```

```python
# ------------------------------
# 5. Verification
# ------------------------------
for (name1, p1), (name2, p2) in zip(
    global_model.named_parameters(), loaded_model.named_parameters()
):
    print(name1, torch.allclose(p1, p2))  # Should print True
```
