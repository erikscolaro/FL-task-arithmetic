```python
import torch
import wandb
from fl_task_arithmetic.task import Net
from utils.wandb_utils import load_model_from_wandb, save_model_to_wandb

entity = "aml-fl-project"
project = "fl-task-arithmetic"
group = "run-test"

# ------------------------------
# 1. Create the model
# ------------------------------
global_model = Net()

# Simulate that the model has non-zero weights
for p in global_model.parameters():
    p.data += 1.234

# ------------------------------
# 2. STORE: save the model as an artifact
# ------------------------------
run = wandb.init(
    entity=entity,
    project=project,
    group=group,
    name="server-store",
    id="server-store",
    resume="never",
    mode="online",
)

save_model_to_wandb(
    run=run,
    model=global_model,
    group=group,
)
run.finish()

print("\n--- MODEL SAVED ON WANDB ---\n")

# ------------------------------
# 3. LOAD: create an empty model
# ------------------------------
loaded_model = Net()

# ------------------------------
# 4. LOAD: retrieve model from artifact
# ------------------------------
load_model_from_wandb(
    model=loaded_model, entity=entity, project=project, group=group, version="latest"
)

print("\n--- MODEL LOADED FROM WANDB ---\n")

# ------------------------------
# 5. Verification
# ------------------------------
for (name1, p1), (name2, p2) in zip(
    global_model.named_parameters(), loaded_model.named_parameters()
):
    print(name1, torch.allclose(p1, p2))  # Should print True
```
